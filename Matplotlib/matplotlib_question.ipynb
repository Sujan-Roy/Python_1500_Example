### Setting up series labels
* Display multiple series data on a single graph
    * You can set labels (legends) to indicate what each series data represents
        * **`matplotlib.pyplot.legend(["label name 1", "label name 2",...])`**
        * **label="label name "** can be specified when plotting series data (**`matplotlib.pyplot.plot()`**), **`matplotlib.pyplot.legend()`** (without argument) can be used to set labels

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(-15, 15)

y1 = 1.5 * x + 10

y2 = -1.5 * x + 15

plt.grid(True)

plt.plot(x, y1, color="#aa0000", label="Y = 1.5 * X + 10")

plt.plot(x, y2, color="g", label="Y = -1.5 * X + 15")

# Here is the code to set the series label (legend) display


plt.show()
```

### Adjusting margins around subplots
* When subplots are placed in a figure, the name or title of an axis may overlap with the neighboring subplot
    * This can be avoided by adjusting the margins around the subplots
    * **`matplotlib.pyplot.subplots_adjust(wspace=percentage of horizontal spacing, hspace=percentage of vertical spacing)`**
        * Percentages are based on the vertical and horizontal axes of the graph

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


x = np.linspace(-15, 15)

y = 1.5 * x + 10

fig = plt.figure(figsize=(10, 5))

ax = fig.add_subplot(1, 3, 2)

# Here is the code to space the horizontal spacing of the subplots by a ratio of 0.75


ax.plot(x, y)

for i in range(3):
    if i == 1:.
        continue
    else:
        fig.add_subplot(1, 3, i+1)

plt.show()
```

### Setting the display range of a graph in a subplot
* Set the display range of a graph in units of subplots
    * Set for a subplot object (**ax**)
        * Specify range of X axis: **`ax.set_xlim([min, max])`**
        * range specification for Y axis: **`ax.set_ylim([min, max])`**
* Note that the method of specifying the range is different from that of setting the display range of a graph when no subplot is used.

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline


x = np.linspace(-15, 15)

y = 1.5 * x + 10

fig = plt.figure(figsize=(10, 5))

ax = fig.add_subplot(1, 3, 2)

plt.subplots_adjust(wspace = 0.75)

# Here is the code to set the display range of the y-axis of the subplot ax to [10,20


ax.plot(x, y)

for i in range(3):
    if i == 1:
        continue
    else:
        fig.add_subplot(1, 3, i+1)

plt.show()
```

### Display grid on graphs in subplots
* Graphs in subplots (ax) also have grid (lattice) display hidden by default
* Grid display: **`ax.grid(True)`**

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(-15, 15)

y = 1.5 * x + 10

fig = plt.figure(figsize=(10, 5))

ax = fig.add_subplot(1, 3, 2)

plt.subplots_adjust(wspace = 0.75)


ax.set_title("y = 1.5 * x + 10")

ax.set_xlabel("X")

ax.set_ylabel("Y")

## Here is the code to set the grid display


ax.plot(x, y)

for i in range(3):
    if i == 1:
        continue
    else:
        fig.add_subplot(1, 3, i+1)

plt.show()
```

### Set ticks on axes of graphs in subplots
* Scales can also be set manually on subplots (ax)
    * Note that the setting method is different from that without subplots
    * Position of ticks to be inserted on the x-axis
        * **`ax.set_xticks(list of insertion points)`**
    * Information on labels to be inserted on the X axis
        * **`ax.set_xticklabels(list of ticks)`**
    * ** position of ticks to insert on Y-axis.
        * **`ax.set_yticks(list of insertion points)`**
    * information of labels to be inserted on the Y-axis
        * **`ax.set_yticklabels(list of ticks)`**

```python
import matplotlib.pyplot as pltimport numpy as np

%matplotlib inline

x = np.linspace(-15, 15)

y = 1.5 * x + 10

fig = plt.figure(figsize=(10, 5))

ax = fig.add_subplot(1, 3, 2)

plt.subplots_adjust(wspace = 0.75)


ax.set_title("y = 1.5 * x + 10")

ax.set_xlabel("X")

ax.set_ylabel("Y")

## Here is the code to specify the X-axis tick position

ax.set_xticklabels(["▲10", "0", "10"])

ax.grid(True)

ax.plot(x, y)

for i in range(3):
    if i == 1:
        continue
    else:
        fig.add_subplot(1, 3, i+1)

plt.show()
```
